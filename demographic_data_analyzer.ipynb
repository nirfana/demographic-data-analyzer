##### What is the percentage of people who have a Bachelor's degree?
there's 2 way on how to find the percentage, I'll break it down on the code column

```python
# 1. make definition for better understanding
bachelor_education_counts = len(df[df['education'] == "Bachelors"])
education_counts = len(df['education'])
same_percentage_bachelors = round(bachelor_education_counts / education_counts * 100, 1)
print(same_percentage_bachelors)
```

```python
# 2. make a simple line 
percentage_bachelors = round(len(df[df['education'] == "Bachelors"]) / len(df['education'])*100, 1)
print(percentage_bachelors)
```

##### What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
##### What percentage of people without advanced education make more than 50K?

with `Bachelors`, `Masters`, or `Doctorate`

```python
higher_education = df[df['education'].isin(["Bachelors", "Masters", "Doctorate"])]
higher_education
```

without `Bachelors`, `Masters`, or `Doctorate`

```python
lower_education = df[~df['education'].isin(["Bachelors", "Masters", "Doctorate"])]
lower_education
```

##### percentage with salary >50K
There's aslo 2 way to find the percentage, the first one I made definition on each and the second one is keeping the code within a line

first

```python
x_higher = len(higher_education[higher_education.salary == ">50K"])
y_higher = len(higher_education)

xy_higher_education_rich = round(x_higher/y_higher * 100, 1)
xy_higher_education_rich
```

```python
x_lower = len(lower_education[lower_education.salary == ">50K"])
y_lower = len(lower_education)

xy_lower_education_rich = round(x_lower/y_lower * 100, 1)
xy_lower_education_rich
```

second way, i love keeping my code on single line 

```python
higher_education_rich = round(len(higher_education[higher_education.salary == ">50K"]) / len(higher_education)*100, 1)
higher_education_rich
```

```python
lower_education_rich = round(len(lower_education[lower_education.salary == ">50K"]) / len(lower_education)*100, 1)
lower_education_rich
```

##### What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

#####  What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = df[df['hours-per-week'] == min_work_hours]
num_min_workers
```

```python
rich_percentage = round(len(num_min_workers[num_min_workers.salary == ">50K"]) / len(num_min_workers) * 100, 1)
rich_percentage
```

##### What country has the highest percentage of people that earn >50K?

```python
highest_earning_country = (df[df['salary'] == ">50K"]['native-country'].value_counts() / df['native-country'].value_counts() * 100 ).idxmax()
highest_earning_country
```

##### Identify the most popular occupation for those who earn >50K in India.

```python
IN_occupation = df[df['native-country'] == "India"]['occupation'].value_counts()
IN_occupation
```

```python
top_IN_occupation = df[df['native-country'] == "India"]['occupation'].value_counts().idxmax()
top_IN_occupation
```
